8.2 함수의 그래프 그리기

```python
import numpy as np
```

```python
from matplotlib import pyplot as plt
```

```python
x=np.arange(-10,10,1)
for a in [2,-2]:
  plt.plot(x,a*x, label = 'y = {}x'.format(a))
plt.grid()
plt.legend(fontsize=15,loc=8)
```

```python
x=np.arange(-10,10,1)
plt.figure(figsize = (5,7))
for a in [1/2,0,2,8]:
  plt.plot(x,a*x,label='y={}x'.format(a))
plt.grid()
plt.legend(fontsize = 14, loc=4)
```

```python
x=np.arange(-10,10,1)
plt.figure(figsize = (5,7))
for a in [-1/2,0,-2,-8]:
  plt.plot(x,a*x,label='y={}x'.format(a))
plt.grid()
plt.legend(fontsize = 14, loc=4)
```

```python
x=np.arange(-10,10)
b=np.arange(-5,5,2)
for i in b:
  plt.plot(x,x+i,label='y = x+{}'.format(i))
plt.grid()
plt.legend(fontsize=13)
```

```python
a=1
x=np.arange(-10,11,1)
plt.figure(figsize=(5,8))
plt.scatter(x,1/x)
plt.axvline(x=0,c='k')
plt.axhline(y=0,c='k')
plt.grid()
```

```python
a=1
x=np.arange(-10,11,1)
plt.figure(figsize=(5,8))
plt.scatter(x,-2/x)
plt.axvline(x=0,c='k')
plt.axhline(y=0,c='k')
plt.grid()
```

```python
a=2
x=np.arange(-10,10,1)
p=np.arange(-5,6,2)
plt.figure(figsize=(5,5))
for i in p:
  plt.plot(x,a*(x-i),label = 'shift {}'.format(i))
plt.grid()
plt.legend()
```

8.3 히스토그램

```python
weight = 22,24,26,30,32,40,35,45,20,29,34,36,36,38,39,48,43,37,33,31,29,39,26,29
bins = np.arange(20,55,5)
hist,bins = np.histogram(weight,bins)
print(hist)
print(bins)
```

```python
total_number = len(weight)
```

```python
hist_normal = np.asarray(hist)/total_number
print(hist_normal)
```

```python
sum_hist_normal=sum(hist_normal)
print(sum_hist_normal)
```

```python
accumul_weight=[]
previous=0
for i in hist:
  previous += i
  accumul_weight.append(previous)
print(accumul_weight)
```

```python
plt.hist(weight,bins)
plt.xlabel('Weight (kg)',fontsize = 14)
plt.xticks(fontsize = 14)
plt.yticks(fontsize = 14)
```

```python
plt.hist(weight,bins, rwidth = 0.8)
plt.xlabel('Weight (kg)',fontsize = 14)
plt.xticks(fontsize = 14)
plt.yticks(fontsize = 14)
```

```python
plt.hist(weight,bins, rwidth = 0.8,color='green')
plt.xlabel('Weight (kg)',fontsize = 14)
plt.xticks(fontsize = 14)
plt.yticks(fontsize = 14)
```

```python
plt.hist(weight,bins, rwidth = 0.8,color='green',alpha=0.5)
plt.grid()
plt.xlabel('Weight (kg)',fontsize = 14)
plt.xticks(fontsize = 14)
plt.yticks(fontsize = 14)
```

```python
x_label=['22.5','27.5','32.5','37.5','42.5','47.5']
plt.bar(x_label, hist)
plt.grid()
```
